```python
def remove_outliers_iqr(df, columns, factor=1.5):

  for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - factor * IQR
    upper_bound = Q3 + factor * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
  return df
```

```python
#Seprating
```

```python
X = df.drop('Potability',axis=1)
y = df['Potability']
```

```python
#Splitting
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
#Scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,classification_report , accuracy_score
```

```python
Model = DecisionTreeClassifier(max_depth=11,min_samples_split=6,min_samples_leaf =6,criterion='gini',random_state = 1)
Model.fit(X_train,y_train)
```

```python
y_pred = Model.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
unique_labels = y.unique()
labels_as_strings = [str(label) for label in unique_labels]
class_names_str = labels_as_strings
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt


plt.figure(figsize=(20, 15))
plot_tree(Model, filled=True, feature_names=X.columns,
          class_names=class_names_str,  # Use string class names
          rounded=True)
plt.show()
```
